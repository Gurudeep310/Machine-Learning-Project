# Content Based Recommender System

```python
import pandas as pd
```

```python
df = pd.read_excel('MaiData.xlsx')
df.drop(columns = ['Runtime','IMDB','Language','Director','Actor','Actress','Year','Budget','BoxOffice','Review(5-best,4-Good,3-ontimewatch,2-worst,1-dontwatch)'], inplace = True)
df2 = df.copy()
for i in range(df.shape[0]):
    df['user'] = 'Gurudeep'
first_column = df.pop('user')
df.insert(0, 'user', first_column)
second_column = df.pop('MyRating')
df.insert(2, 'MyRating', second_column)
df.head(2)
```

```python
df1 = df
classes = list(df1.columns)[1:]
classes = classes[3:]
for cla in classes:
    df1[cla] = (df1[cla]*df1['MyRating'])
df1.head(3)
```

```python
user_profiles = df1.drop('MyRating', axis = 1)
user_profiles = user_profiles.groupby(['user']).sum()
user_profiles["total"] = user_profiles[classes].sum(axis=1)
user_profiles
```

```python
for cla in classes:
    user_profiles[cla] = (user_profiles[cla] / user_profiles["total"])
user_profiles = user_profiles.drop("total", axis=1)
user_profiles
```

```python
df.head(3)
```

```python
df2
```

```python
import numpy as np
categories_np =np.array(df2.drop(columns = ["MovieName","MyRating"], axis=1))
categories_np
```

```python
user_profiles_np = np.array(user_profiles)
user_profiles_np
```

```python
preferences = np.dot(user_profiles_np, categories_np.T)
min_rating = 2
max_rating = 10
preferences = preferences * (max_rating - min_rating) + min_rating
preferences
```

```python
predictions = df[['user','MovieName','MyRating']]
predictions["predicted_rating"] = np.ravel(preferences)
predictions.head()
```

```python
predictions["squared_error"] = (predictions["predicted_rating"] - predictions["MyRating"]) ** 2
predictions.head(2)
```

```python
predictions["squared_error"].mean() ** .5    # High Error 
```
